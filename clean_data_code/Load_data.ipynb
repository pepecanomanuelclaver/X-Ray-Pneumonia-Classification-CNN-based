```python
# Attempt 1, doesnt work for every json file, since they dont all have all info
def make_json_to_df_row(json_path):
    with open(json_path) as json_file:
        data = json.load(json_file)
    temp_df = pd.DataFrame.from_dict(data.get("annotations")).T.reset_index()
    print("tem", len(temp_df.columns))
    temp_df_2 = pd.DataFrame.from_dict([data.get("image")])
    if len(temp_df.columns) > 10:
        pandas_row = pd.concat([temp_df_2[["seq","filename"]] ,temp_df[[6,7,8,10]]], axis=1, ignore_index =True).head(1)
        pandas_row2 = pd.concat([pandas_row, temp_df[[0,1]].iloc[1]], axis=1, ignore_index = True)
        one_row = pd.concat([pandas_row2.head(1), pandas_row2[[6]].iloc[1].reset_index()], axis=1, ignore_index = True).drop(7, axis=1)
        final_row = one_row.set_index([1])
        final_row = final_row.rename(columns={0: 'seq', 2:"Type", 3:"Pneumonia", 4:"Covid", 5: "Patient_id", 6:"Lung1", 8:"Lung2"})
        final_row = final_row.rename_axis('Filename')
        return final_row
    else:
        print("not complete")
        return True
    
```

```python
# Thought there were only two types of completeness but there are more so this becomes very hard code if you want to
# Address all options and append issues
def make_json_to_df_row3(filepath):
    with open(filepath) as json_file:
        data = json.load(json_file)
    temp_df = pd.DataFrame.from_dict(data.get("annotations")).T.reset_index()
    temp_df_2 = pd.DataFrame.from_dict([data.get("image")])
    temp_df_2 = temp_df_2[["filename"]]
    final_df = pd.concat([temp_df, temp_df_2], axis =1)
    lungs = pd.concat([temp_df.T[[1]].iloc[1], temp_df.T[[1]].iloc[2]], axis =1).reset_index()
    lungs = lungs.rename(columns = {'index': 100, 0:"lung1", 1:"lung2"})
    testing = pd.concat([final_df.head(1), lungs], axis = 1)
    first_column = testing.pop('filename')
    testing.insert(0, 'filename', first_column)
    testing = testing.drop([100, "index",0,1],axis=1)
#     print("collen",len(testing.columns))
    if len(testing.columns) == 4:
        testing.rename(columns = {2:"type"}, inplace = True)
    elif len(testing.columns) > 16:
        testing = testing.rename(columns = {9:"type"})
        testing = testing[['filename', 'type', 'lung1', 'lung2',2,3,4,5,6,7,8,10,11,12,13,14,15]]
        testing =testing.drop([3,4,5,6,7,8,11,12,12,13,14,15], axis =1)
        testing.rename(columns = {2:"view", 10:"covid"}, inplace = True)
    return testing
```

```python
# Working, might need some extra collumns but can handle different json lengths etc.
def better_df_from_json(json_path):
    df = pd.DataFrame()
    with open(json_path) as json_file:
        data = json.load(json_file)
    filename = data.get("image").get("filename")
    ogfilename = data.get("image").get("original_filename")
    temp = pd.DataFrame.from_dict(data.get("annotations"))
    df['ogfilename'] = [ogfilename]
    df['filename'] = [filename]
    types = re.search("\s*(\S+) Pneumonia", temp['name'].to_string())
    if types == None:
        df['type'] = np.nan
    else:
        df['type'] = [types.group()]
#     df['type'] = [re.search("\s*(\S+) Pneumonia", temp['name'].to_string()).group()]
    if 'polygon' in temp.columns:
        lung = temp['polygon'].iloc[0]
        df['lung1'] = [temp['polygon'].iloc[0]]
        df['lung2'] = [temp['polygon'].iloc[1]]
    else:
        df['lung1'] = np.nan
        df['lung2'] = np.nan
    view = re.search("View\s*(\S+)", temp['name'].to_string())
    covid = re.search("COVID\s*(\S+)", temp['name'].to_string())
    if view == None:
#         print("test")
        df['view'] = np.nan
    else:
        df['view'] = [view.group()]
    if covid == None:
        df['Covid'] = np.nan
    else:
        df['Covid'] = [covid.group()]
    width = filename = data.get("image").get("width")
    height = filename = data.get("image").get("height")
    df["width"] = width
    df["height"] = height
#     df['json_filename'] = json_path
    return df
```

```python
# Get all jsons into df
def all_json_to_df(filepath):
    df = pd.DataFrame(columns=["filename", 'type'])
    entries = os.listdir(filepath)
    for entry in entries:
#         print("entry",entry)
        row = better_df_from_json(filepath+entry)
        row['json_filename'] = entry
        df = pd.concat([df,row])
    return df
```
